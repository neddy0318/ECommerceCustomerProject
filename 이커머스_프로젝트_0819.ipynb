```python
def check_missing_values(df, df_name):
    print(f"{df_name} 결측치 정보:")
    missing_values = df.isnull().sum()  # 결측치 개수
    missing_percentage = (df.isnull().sum() / len(df)) * 100  # 결측치 비율

    missing_df = pd.DataFrame({'결측치 개수': missing_values, '결측치 비율 (%)': missing_percentage})
    missing_df = missing_df[missing_df['결측치 개수'] > 0]
    print(missing_df)
    print("\n")

# 각 데이터프레임에 대해 결측치 확인
check_missing_values(df1, "df1")
check_missing_values(df2, "df2")
check_missing_values(df3, "df3")
check_missing_values(df4, "df4")
check_missing_values(df5, "df5")
```

```python
date_ranges = {
    "df1": ("2019-10-01", "2019-10-31 23:59:59"),
    "df2": ("2019-11-01", "2019-11-30 23:59:59"),
    "df3": ("2019-12-01", "2019-12-31 23:59:59"),
    "df4": ("2020-01-01", "2020-01-31 23:59:59"),
    "df5": ("2020-02-01", "2020-02-29 23:59:59")
}

def check_event_time_outliers(df, df_name, start_date, end_date):
    print(f"\n{df_name} - event_time의 최소값: {df['event_time'].min()}")
    print(f"{df_name} - event_time의 최대값: {df['event_time'].max()}")

    abnormal_times = df[(df['event_time'] < start_date) | (df['event_time'] > end_date)]
    if not abnormal_times.empty:
        print(f"{df_name} - 비정상적인 event_time 값:")
        print(abnormal_times)
    else:
        print(f"{df_name} - 비정상적인 event_time 값 없음.")

check_event_time_outliers(df1, "df1", *date_ranges["df1"])
check_event_time_outliers(df2, "df2", *date_ranges["df2"])
check_event_time_outliers(df3, "df3", *date_ranges["df3"])
check_event_time_outliers(df4, "df4", *date_ranges["df4"])
check_event_time_outliers(df5, "df5", *date_ranges["df5"])
```

```python
def session_duration_analysis(df, df_name):
    session_durations = df.groupby('user_session').agg(
        session_start=('event_time', 'min'),
        session_end=('event_time', 'max'),
        session_length=('event_time', lambda x: x.max() - x.min())
    )

    print(f"\n{df_name} - 세션 길이 통계량:")
    print(session_durations['session_length'].describe())

    short_sessions = session_durations[session_durations['session_length'] < pd.Timedelta(seconds=1)]
    long_sessions = session_durations[session_durations['session_length'] > pd.Timedelta(hours=24)]

    print(f"{df_name} - 비정상적으로 짧은 세션들:")
    print(short_sessions)

    print(f"{df_name} - 비정상적으로 긴 세션들:")
    print(long_sessions)

session_duration_analysis(df1, "df1")
session_duration_analysis(df2, "df2")
session_duration_analysis(df3, "df3")
session_duration_analysis(df4, "df4")
session_duration_analysis(df5, "df5")
```

```python
# 세션 길이 분
def session_duration_analysis(df, df_name):
    session_durations = df.groupby('user_session').agg(
        session_start=('event_time', 'min'),
        session_end=('event_time', 'max'),
        session_length=('event_time', lambda x: x.max() - x.min())
    )

    print(f"\n{df_name} - 세션 길이 통계량:")
    print(session_durations['session_length'].describe())

    short_sessions = session_durations[session_durations['session_length'] < pd.Timedelta(seconds=1)]
    long_sessions = session_durations[session_durations['session_length'] > pd.Timedelta(hours=24)]

    very_long_sessions = session_durations[session_durations['session_length'] > pd.Timedelta(days=30)]

    print(f"{df_name} - 비정상적으로 짧은 세션들:")
    print(short_sessions)

    print(f"{df_name} - 비정상적으로 긴 세션들:")
    print(long_sessions)

    print(f"{df_name} - 한 달 넘는 세션들:")
    print(very_long_sessions)

session_duration_analysis(df1, "df1")
session_duration_analysis(df2, "df2")
session_duration_analysis(df3, "df3")
session_duration_analysis(df4, "df4")
session_duration_analysis(df5, "df5")
```

```python
# 논리적 패턴 분석

def logical_pattern_analysis(df, df_name):
    out_of_order_sessions = []

    grouped = df.groupby('user_session')
    for session_id, group in grouped:
        events = group['event_type'].tolist()

        if 'purchase' in events:
            purchase_index = events.index('purchase')
            if 'view' in events[purchase_index + 1:] or 'cart' in events[purchase_index + 1:]:
                out_of_order_sessions.append(session_id)

    print(f"\n{df_name} - 논리적 패턴이 맞지 않는 세션들:")
    if out_of_order_sessions:
        print(out_of_order_sessions)
    else:
        print("논리적 패턴이 맞지 않는 세션 없음.")

logical_pattern_analysis(df1, "df1")
logical_pattern_analysis(df2, "df2")
logical_pattern_analysis(df3, "df3")
logical_pattern_analysis(df4, "df4")
logical_pattern_analysis(df5, "df5")
```

```python
# 세션 내 사용자 ID와의 관계 확인

def check_user_id_consistency(df, df_name):
    inconsistent_sessions = df.groupby('user_session')['user_id'].nunique()
    inconsistent_sessions = inconsistent_sessions[inconsistent_sessions > 1]

    print(f"\n{df_name} - 서로 다른 user_id를 가진 세션들:")
    if not inconsistent_sessions.empty:
        print(inconsistent_sessions)
    else:
        print("서로 다른 user_id를 가진 세션 없음.")

check_user_id_consistency(df1, "df1")
check_user_id_consistency(df2, "df2")
check_user_id_consistency(df3, "df3")
check_user_id_consistency(df4, "df4")
check_user_id_consistency(df5, "df5")
```

```python
# 세션당 이벤트 수 분석
def session_event_count_analysis(df, df_name):
    event_counts = df.groupby('user_session').size()

    print(f"\n{df_name} - 세션당 이벤트 수 통계량:")
    print(event_counts.describe())

    threshold = event_counts.quantile(0.99)
    outlier_sessions = event_counts[event_counts > threshold]

    print(f"{df_name} - 비정상적으로 많은 이벤트가 발생한 세션들:")
    print(outlier_sessions)

session_event_count_analysis(df1, "df1")
session_event_count_analysis(df2, "df2")
session_event_count_analysis(df3, "df3")
session_event_count_analysis(df4, "df4")
session_event_count_analysis(df5, "df5")
```

```python
# 세션 시간의 논리적 일관성

def check_session_time_consistency(df, df_name):
    inconsistent_sessions = []

    grouped = df.groupby('user_session')
    for session_id, group in grouped:
        times = group['event_time']
        if not times.is_monotonic_increasing:
            inconsistent_sessions.append(session_id)

    print(f"\n{df_name} - 이벤트 시간의 논리적 일관성이 맞지 않는 세션들:")
    if inconsistent_sessions:
        print(inconsistent_sessions)
    else:
        print("이벤트 시간의 논리적 일관성이 맞지 않는 세션 없음.")

check_session_time_consistency(df1, "df1")
check_session_time_consistency(df2, "df2")
check_session_time_consistency(df3, "df3")
check_session_time_consistency(df4, "df4")
check_session_time_consistency(df5, "df5")
```

```python
# 데이터 확인
SELECT *
FROM cosmetics_logs;


SELECT *
FROM cosmetics_logs
WHERE price < 0;


SELECT *
FROM cosmetics_logs
WHERE user_id = '410016187';


SELECT *
FROM cosmetics_logs
WHERE user_id = 410016187 AND product_id = 5670257;

SELECT *
FROM cosmetics_logs
WHERE user_id = 410016187 AND price = -15.87;





SELECT *
FROM cosmetics_logs
LIMIT 10;


# 총 데이터 확인
# 총 rows = 3,533,286
SELECT COUNT(*) AS total_rows
FROM cosmetics_logs;


# 이름이 이상하게 나와서 변경
ALTER TABLE cosmetics_logs RENAME COLUMN 癤�event_time TO event_time;


# Null값 찾기 => null이 아니라 빈 문자열 혹은 공백 문자열이 포함된 것인지 값이 정확하게 나오지 않음.
SELECT
    SUM(CASE WHEN event_time IS NULL THEN 1 ELSE 0 END) AS event_time_nulls,
    SUM(CASE WHEN event_type IS NULL THEN 1 ELSE 0 END) AS event_type_nulls,
    SUM(case when product_id is null then 1 else 0 end) as product_id_nulls,
    SUM(CASE WHEN category_code IS NULL THEN 1 ELSE 0 END) AS category_code_nulls,
    SUM(CASE WHEN brand IS NULL THEN 1 ELSE 0 END) AS brand_nulls,
    SUM(CASE WHEN price IS NULL THEN 1 ELSE 0 END) AS price_nulls,
    SUM(CASE WHEN user_id IS NULL THEN 1 ELSE 0 END) AS user_id_nulls,
    SUM(CASE WHEN user_session IS NULL THEN 1 ELSE 0 END) AS user_session_nulls
FROM cosmetics_logs;


# category_code, brand에는 널값이 있는 것이 확실하기에 체크
# category_code = 3,473,821(98.32%) / brand_nulls = 1,510,289(42.74%)
#brand에서 null값을 Nobrand로 바꿔보는 것도..?
SELECT
    SUM(CASE WHEN event_time IS NULL THEN 1 ELSE 0 END) AS event_time_nulls,
    SUM(CASE WHEN event_type IS NULL THEN 1 ELSE 0 END) AS event_type_nulls,
    SUM(CASE WHEN category_code IS NULL OR category_code = '' OR TRIM(category_code) = '' THEN 1 ELSE 0 END) AS category_code_nulls,
    SUM(CASE WHEN brand IS NULL OR brand = '' OR TRIM(brand) = '' THEN 1 ELSE 0 END) AS brand_nulls,
    SUM(CASE WHEN price IS NULL THEN 1 ELSE 0 END) AS price_nulls,
    SUM(CASE WHEN user_id IS NULL THEN 1 ELSE 0 END) AS user_id_nulls,
    SUM(CASE WHEN user_session IS NULL THEN 1 ELSE 0 END) AS user_session_nulls
FROM cosmetics_logs;


# view 1,728,331
# cart 927,124(53.64%)
# remove_from_cart 664,655(71.69%)
# purchase 213,176(22.99%)  / 물건을 본 사람 중에서는 12.33% / 보통 상품별로 이정도가 괜찮은 수치인지가 궁금
SELECT event_type, COUNT(*) AS ecent_type_category_count
FROM cosmetics_logs
GROUP BY event_type;


# 총 유저수 370,154
SELECT COUNT(DISTINCT user_id) AS unique_user_count
FROM cosmetics_logs;


# 일단 brand컬럼 빈값 unknown_brand로 바꿔버리기
START TRANSACTION;
UPDATE cosmetics_logs
SET brand = 'unknown_brand'
WHERE brand IS NULL OR brand = '' OR TRIM(brand) = '';
COMMIT;


# brand 각 몇개일까?
# 판매량 TOP5
# unknown_brand 91,695
# 1. runail 18,199
# 2. irisk 10,583
# 3. grattol 8,171
# 4. bpw.style 7,014
# 5. masura 6,985
SELECT brand, COUNT(*) AS count
FROM cosmetics_logs
WHERE event_type = 'purchase'
GROUP BY brand
ORDER BY count DESC;

# 관심 TOP5
# unknown_brand 91,695
# 1. runail 108,097
# 2. grattol 83,257
# 3. irisk 66,529
# 4. masura 50,363
# 5. kapous 45,225
SELECT brand, COUNT(*) AS count
FROM cosmetics_logs
WHERE event_type = 'view'
GROUP BY brand
ORDER BY count DESC;

# 장바구니 TOP5
# unknown_brand 403,305
# 1. runail 80,411
# 2. irisk 45,841
# 3. grattol 37,841
# 4. masura 33,788
# 5. bpw.style 21,995
SELECT brand, COUNT(*) AS count
FROM cosmetics_logs
WHERE event_type = 'cart'
GROUP BY brand
ORDER BY count DESC;

# 장바구니 제거 TOP5
# unknown_brand 296,236
# 1. runail 49,510
# 2. masura 34,049
# 3. irisk 29,565
# 4. grattol 26,651
# 5. bpw.style 18,014
SELECT brand, COUNT(*) AS count
FROM cosmetics_logs
WHERE event_type = 'remove_from_cart'
GROUP BY brand
ORDER BY count DESC;



# 브랜드별 총 판매량과 갯수
# 매출 TOP5
# 1. NO_Brand 440,207.57
# 2. runail 58,177.26
# 3. grattol 43,793.50
# 4. irisk 35,291.51
# 5. uno 29,084.76
SELECT brand, COUNT(*) as sales_count, SUM(price) as total_revenue
FROM cosmetics_logs
WHERE event_type = 'purchase'
GROUP BY brand
ORDER BY total_revenue DESC;


# 브랜드별 평균 가격
SELECT brand, AVG(price) AS average_price
FROM cosmetics_logs
GROUP BY brand
ORDER BY average_price DESC;

# 종합
SELECT
    b.brand,
    p.purchase_count,
    v.view_count,
    c.cart_count,
    rfc.remove_from_cart_count,
    pr.sales_count,
    pr.total_revenue,
    ap.average_price
FROM
    (SELECT DISTINCT brand FROM cosmetics_logs) b
LEFT JOIN
    (SELECT brand, COUNT(*) AS purchase_count
     FROM cosmetics_logs
     WHERE event_type = 'purchase'
     GROUP BY brand) p ON b.brand = p.brand
LEFT JOIN
    (SELECT brand, COUNT(*) AS view_count
     FROM cosmetics_logs
     WHERE event_type = 'view'
     GROUP BY brand) v ON b.brand = v.brand
LEFT JOIN
    (SELECT brand, COUNT(*) AS cart_count
     FROM cosmetics_logs
     WHERE event_type = 'cart'
     GROUP BY brand) c ON b.brand = c.brand
LEFT JOIN
    (SELECT brand, COUNT(*) AS remove_from_cart_count
     FROM cosmetics_logs
     WHERE event_type = 'remove_from_cart'
     GROUP BY brand) rfc ON b.brand = rfc.brand
LEFT JOIN
    (SELECT brand, COUNT(*) AS sales_count, SUM(price) AS total_revenue
     FROM cosmetics_logs
     WHERE event_type = 'purchase'
     GROUP BY brand) pr ON b.brand = pr.brand
LEFT JOIN
    (SELECT brand, AVG(price) AS average_price
     FROM cosmetics_logs
     GROUP BY brand) ap ON b.brand = ap.brand
ORDER BY
    pr.total_revenue DESC;





# 사용자별 평균 이벤트 수
SELECT user_id, COUNT(*) AS event_count
FROM cosmetics_logs
GROUP BY user_id
ORDER BY event_count DESC;

SELECT user_id, COUNT(*) AS count
FROM cosmetics_logs
WHERE event_type = 'purchase'
GROUP BY user_id
ORDER BY count DESC;

SELECT user_id, COUNT(*) AS count
FROM cosmetics_logs
WHERE event_type = 'view'
GROUP BY user_id
ORDER BY count DESC;

SELECT user_id, COUNT(*) AS count
FROM cosmetics_logs
WHERE event_type = 'cart'
GROUP BY user_id
ORDER BY count DESC;

SELECT user_id, COUNT(*) AS count
FROM cosmetics_logs
WHERE event_type = 'remove_from_cart'
GROUP BY user_id
ORDER BY count DESC;



# 유저별 평균 페이지 뷰
SELECT user_id, AVG(page_views) AS avg_page_views_before_purchase
FROM (
    SELECT user_id, user_session, COUNT(*) AS page_views
    FROM cosmetics_logs
    WHERE event_type = 'view'
    GROUP BY user_id, user_session
) AS session_views
GROUP BY user_id
ORDER BY avg_page_views_before_purchase DESC;

# 유저별 구매 전 평균 페이지 조회 수
WITH session_purchases AS (
    SELECT
        user_id,
        user_session,
        MIN(STR_TO_DATE(SUBSTRING_INDEX(event_time, ' UTC', 1), '%Y-%m-%d %H:%i:%s')) AS purchase_time
    FROM
        cosmetics_logs
    WHERE
        event_type = 'purchase'
    GROUP BY
        user_id, user_session
),
session_views_before_purchase AS (
    SELECT
        v.user_id,
        v.user_session,
        COUNT(*) AS page_views
    FROM
        cosmetics_logs v
    LEFT JOIN
        session_purchases p ON v.user_id = p.user_id AND v.user_session = p.user_session
    WHERE
        v.event_type = 'view' AND
        (p.purchase_time IS NULL OR STR_TO_DATE(SUBSTRING_INDEX(v.event_time, ' UTC', 1), '%Y-%m-%d %H:%i:%s') < p.purchase_time)
    GROUP BY
        v.user_id, v.user_session
)
SELECT
    user_id,
    AVG(page_views) AS avg_page_views_before_purchase
FROM
    session_views_before_purchase
GROUP BY
    user_id
ORDER BY
    avg_page_views_before_purchase DESC;

# 시간대별 이벤트 분포
SELECT DATE_FORMAT(STR_TO_DATE(SUBSTRING_INDEX(event_time, ' UTC', 1), '%Y-%m-%d %H:%i:%s'), '%Y-%m-%d %H:00:00') AS hour,
       event_type,
       COUNT(*) AS event_count
FROM cosmetics_logs
GROUP BY hour, event_type
ORDER BY hour, event_type;




# 가장 높은 평균 가격의 브랜드
# strong 204.02
-- SELECT brand, AVG(price) AS average_price
-- FROM cosmetics_logs
-- GROUP BY brand
-- ORDER BY average_price DESC
-- LIMIT 1;


# 가장 낮은 평균 가격의 브랜드
# soleo 0.86
-- SELECT brand, AVG(price) AS average_price
-- FROM cosmetics_logs
-- GROUP BY brand
-- ORDER BY average_price ASC
-- LIMIT 1;

# product_id 엄청 많음 ...
-- SELECT product_id, COUNT(*) AS count
-- FROM cosmetics_logs
-- GROUP BY product_id
-- ORDER BY count DESC;


# 총 Product_id : 44,624
-- SELECT COUNT(DISTINCT product_id) AS unique_product_count
-- FROM cosmetics_logs;


# 시간 관련 파생변수 생성
SELECT *,
       HOUR(STR_TO_DATE(SUBSTRING_INDEX(event_time, ' UTC', 1), '%Y-%m-%d %H:%i:%s')) AS hour_of_day,
       DAYOFWEEK(STR_TO_DATE(SUBSTRING_INDEX(event_time, ' UTC', 1), '%Y-%m-%d %H:%i:%s')) - 1 AS day_of_week, -- 월요일을 0으로 설정
       WEEK(STR_TO_DATE(SUBSTRING_INDEX(event_time, ' UTC', 1), '%Y-%m-%d %H:%i:%s')) AS week_of_year,
       MONTH(STR_TO_DATE(SUBSTRING_INDEX(event_time, ' UTC', 1), '%Y-%m-%d %H:%i:%s')) AS month
FROM cosmetics_logs;


# 사용자 행동 관련 파생 변수
-- user_id: 사용자 ID
-- unique_products_viewed: 사용자가 조회한 고유한 제품 수
-- unique_products_in_cart: 사용자가 장바구니에 추가한 고유한 제품 수
-- unique_products_removed_from_cart: 사용자가 장바구니에서 제거한 고유한 제품 수
-- unique_products_purchased: 사용자가 구매한 고유한 제품 수
-- products_in_cart: 장바구니에 추가한 제품 수 (중복 포함)
-- total_spent: 총 지출 금액
WITH user_data AS (
    SELECT
        user_id,
        COUNT(DISTINCT CASE WHEN event_type = 'view' THEN product_id END) AS unique_products_viewed,
        COUNT(DISTINCT CASE WHEN event_type = 'cart' THEN product_id END) AS unique_products_in_cart,
        COUNT(DISTINCT CASE WHEN event_type = 'remove_from_cart' THEN product_id END) AS unique_products_removed_from_cart,
        COUNT(DISTINCT CASE WHEN event_type = 'purchase' THEN product_id END) AS unique_products_purchased,
        COUNT(CASE WHEN event_type = 'cart' THEN product_id END) AS products_in_cart,
        SUM(CASE WHEN event_type = 'purchase' THEN price ELSE 0 END) AS total_spent
    FROM
        cosmetics_logs
    GROUP BY
        user_id
)
SELECT
    user_id,
    unique_products_viewed,
    unique_products_in_cart,
    unique_products_removed_from_cart,
    unique_products_purchased,
    products_in_cart,
    total_spent
FROM
    user_data;

# 가격 수리통계  => mysql에는 중앙값, 사분위수 구하는 함수 없어서 파이썬 써봐야할듯 .. / 가격에 -가 있음!!
SELECT
    AVG(price) AS mean_price,
    MIN(price) AS min_price,
    MAX(price) AS max_price,
    STDDEV(price) AS stddev_price
FROM
    cosmetics_logs;


# 가격 범주 컬럼 만들기
# 가격이 0 이하인 값은 0으로 처리
UPDATE cosmetics_logs
SET price = 0
WHERE price < 0;

-- 가격 범주 생성
SELECT *,
       CASE
           WHEN price <= 8.87 THEN 'low'
           WHEN price > 8.87 AND price <= 28.73 THEN 'medium'
           ELSE 'high'
       END AS price_category
FROM cosmetics_logs;


# 브랜드 인기도 컬럼 만들기  => 근데 많이 보거나 혹은 장바구니에 담았어도 구매하지 않았다면 그게 인기있는 것인지 .. ?
WITH brand_event_counts AS (
    SELECT
        brand,
        COUNT(*) AS event_count
    FROM
        cosmetics_logs
    WHERE
        brand IS NOT NULL AND brand != ''
    GROUP BY
        brand
),
ranked_brands AS (
    SELECT
        brand,
        event_count,
        NTILE(10) OVER (ORDER BY event_count DESC) AS decile
    FROM
        brand_event_counts
),
top_10_percent_threshold AS (
    SELECT
        MIN(event_count) AS threshold
    FROM
        ranked_brands
    WHERE
        decile = 1
)
SELECT
    c.*,
    CASE
        WHEN b.event_count >= (SELECT threshold FROM top_10_percent_threshold) THEN 'popular'
        ELSE 'not_popular'
    END AS brand_popularity
FROM
    cosmetics_logs c
LEFT JOIN
    brand_event_counts b ON c.brand = b.brand
WHERE
    c.brand IS NOT NULL AND c.brand != '';




# 브랜드 가격대, 인기도 합본
WITH brand_event_counts AS (
    SELECT
        brand,
        COUNT(*) AS event_count
    FROM
        cosmetics_logs
    WHERE
        brand IS NOT NULL AND brand != ''
    GROUP BY
        brand
),
ranked_brands AS (
    SELECT
        brand,
        event_count,
        NTILE(10) OVER (ORDER BY event_count DESC) AS decile
    FROM
        brand_event_counts
),
top_10_percent_threshold AS (
    SELECT
        MIN(event_count) AS threshold
    FROM
        ranked_brands
    WHERE
        decile = 1
)
SELECT
    c.*,
    CASE
        WHEN c.price <= 8.87 THEN 'low'
        WHEN c.price > 8.87 AND c.price <= 28.73 THEN 'medium'
        ELSE 'high'
    END AS price_category,
    CASE
        WHEN b.event_count >= (SELECT threshold FROM top_10_percent_threshold) THEN 'popular'
        ELSE 'not_popular'
    END AS brand_popularity
FROM
    cosmetics_logs c
LEFT JOIN
    brand_event_counts b ON c.brand = b.brand
WHERE
    c.brand IS NOT NULL AND c.brand != '';





# 변수 생성 전 데이터 전처리
CREATE TEMPORARY TABLE user_sessions AS
SELECT
    user_id,
    user_session,
    MIN(STR_TO_DATE(SUBSTRING_INDEX(event_time, ' UTC', 1), '%Y-%m-%d %H:%i:%s')) AS session_start,
    MAX(STR_TO_DATE(SUBSTRING_INDEX(event_time, ' UTC', 1), '%Y-%m-%d %H:%i:%s')) AS session_end,
    TIMESTAMPDIFF(MINUTE,
        MIN(STR_TO_DATE(SUBSTRING_INDEX(event_time, ' UTC', 1), '%Y-%m-%d %H:%i:%s')),
        MAX(STR_TO_DATE(SUBSTRING_INDEX(event_time, ' UTC', 1), '%Y-%m-%d %H:%i:%s'))) AS session_length,
    SUM(CASE WHEN event_type = 'purchase' THEN price ELSE 0 END) AS total_spent_per_session
FROM
    cosmetics_logs
GROUP BY
    user_id, user_session;
# 확인
select *
from user_sessions;

# total_sessions: 사용자의 총 세션 수
CREATE TEMPORARY TABLE user_total_sessions AS
SELECT
    user_id,
    COUNT(*) AS total_sessions
FROM
    user_sessions
GROUP BY
    user_id;
# 확인
select *
from user_total_sessions;


# total_events: 사용자의 총 이벤트 수
CREATE TEMPORARY TABLE user_total_events AS
SELECT
    user_id,
    COUNT(*) AS total_events
FROM
    cosmetics_logs
GROUP BY
    user_id;
# 확인
select *
from user_total_events;


# average_session_length: 사용자의 평균 세션 길이
CREATE TEMPORARY TABLE user_avg_session_length AS
SELECT
    user_id,
    AVG(session_length) AS average_session_length
FROM
    user_sessions
GROUP BY
    user_id;
# 확인
select *
from user_avg_session_length;


# average_spent_per_session: 사용자의 세션당 평균 지출액
CREATE TEMPORARY TABLE user_avg_spent_per_session AS
SELECT
    user_id,
    AVG(total_spent_per_session) AS average_spent_per_session
FROM
    user_sessions
GROUP BY
    user_id;
# 확인
select *
from user_avg_spent_per_session;


# view_to_cart_rate: 조회에서 장바구니로 전환율
CREATE TEMPORARY TABLE user_view_to_cart_rate AS
SELECT
    user_id,
    CASE
        WHEN SUM(CASE WHEN event_type = 'view' THEN 1 ELSE 0 END) = 0
        THEN 0
        ELSE SUM(CASE WHEN event_type = 'cart' THEN 1 ELSE 0 END) / SUM(CASE WHEN event_type = 'view' THEN 1 ELSE 0 END)
    END AS view_to_cart_rate
FROM
    cosmetics_logs
GROUP BY
    user_id;
# 확인
select *
from user_view_to_cart_rate;


# cart_to_purchase_rate: 장바구니에서 구매로 전환율
CREATE TEMPORARY TABLE user_cart_to_purchase_rate AS
SELECT
    user_id,
    CASE
        WHEN SUM(CASE WHEN event_type = 'cart' THEN 1 ELSE 0 END) = 0
        THEN 0
        ELSE SUM(CASE WHEN event_type = 'purchase' THEN 1 ELSE 0 END) / SUM(CASE WHEN event_type = 'cart' THEN 1 ELSE 0 END)
    END AS cart_to_purchase_rate
FROM
    cosmetics_logs
GROUP BY
    user_id;
# 확인
select *
from user_cart_to_purchase_rate;



SELECT
    u.user_id,
    s.total_sessions,
    e.total_events,
    l.average_session_length,
    p.average_spent_per_session,
    v.view_to_cart_rate,
    c.cart_to_purchase_rate
FROM
    (SELECT DISTINCT user_id FROM cosmetics_logs) u
LEFT JOIN
    user_total_sessions s ON u.user_id = s.user_id
LEFT JOIN
    user_total_events e ON u.user_id = e.user_id
LEFT JOIN
    user_avg_session_length l ON u.user_id = l.user_id
LEFT JOIN
    user_avg_spent_per_session p ON u.user_id = p.user_id
LEFT JOIN
    user_view_to_cart_rate v ON u.user_id = v.user_id
LEFT JOIN
    user_cart_to_purchase_rate c ON u.user_id = c.user_id;
```

1. 실루엣 계수 (Silhouette Coefficient)
* 설명: 각 데이터 포인트의 응집도와 분리도를 동시에 고려한 지표입니다. -1에서 1 사이의 값을 가지며, 1에 가까울수록 좋은 군집화 결과를 나타냅니다.
* 계산:
    * a(i):포인트 𝑖와 같은 군집 내의 다른 포인트들 사이의 평균 거리
    * b(i): 포인트 𝑖와 가장 가까운 다른 군집의 포인트들 사이의 평균 거리
    * 실루엣 계수 𝑠(𝑖) = (b(i)−a(i)) / (max(a(i),b(i)))

2. 칼린스키-하라바즈 지수 (Calinski-Harabasz Index)
* 설명: 군집 내의 데이터 포인트들의 분산과 군집 간의 분산을 비율로 나타낸 지표입니다. 값이 클수록 좋은 군집화 결과를 나타냅니다.
* 계산: 군집 간의 거리 제곱합 / 군집 내의 거리 제곱합

3. 던 지수 (Dunn Index)
* 설명: 최소 군집 간 거리와 최대 군집 내 거리의 비율로, 값이 클수록 좋은 군집화 결과를 나타냅니다.
* 계산: 최소 군집 간 거리 / 최대 군집 내 거리

4. Davies-Bouldin Index
* 설명: 군집 내의 응집도와 군집 간의 분리도를 동시에 고려한 지표입니다. 값이 작을수록 좋은 군집화 결과를 나타냅니다.
* 계산: 각 군집의 평균 거리의 합 / 군집 간의 중심 거리

5. 엘보 방법 (Elbow Method)
* 설명: 군집 수에 따른 SSE (Sum of Squared Errors, 오차 제곱합)를 계산하여, SSE가 급격히 감소하다가 완만해지는 지점을 찾는 방법입니다. 완만해지는 지점이 최적의 군집 수를 나타냅니다.
* 계산: 군집 수에 따른 SSE 값을 그래프로 그려서 관찰

6. 변별력 지수 (Discriminant Index)
* 설명: 군집 간의 분산과 군집 내의 분산을 동시에 고려하여, 군집 간의 명확성을 평가합니다.
* 계산: 군집 간의 거리 / 군집 내의 거리

7. 갭 통계량 (Gap Statistic)
* 설명: 실제 데이터와 무작위 데이터를 비교하여 군집 수를 결정합니다. 값이 클수록 좋은 군집 수를 나타냅니다.
* 계산: 관측된 데이터의 Wk와 무작위 데이터의 Wk의 차이

8. BIC (Bayesian Information Criterion)
* 설명: 모델의 복잡성과 적합도를 동시에 고려하여, 최적의 군집 수를 결정합니다. 값이 작을수록 좋은 군집화 결과를 나타냅니다.
* 계산: 모델의 로그 가능도와 모델 복잡도를 고려한 지표

1. 차원 축소 없이 군집 분석

2. 차원 축소 없이 데이터 전처리 후 군집 분석

3. 차원 축소 후 군집 분석

4. 차원 축소 및 데이터 전처리 후 군집 분석

1. K-Means Clustering
* 특징: 가장 널리 사용되는 군집 분석 알고리즘 중 하나입니다. 데이터 포인트를 K개의 군집으로 나누고, 각 군집의 중심(centroid)을 반복적으로 업데이트합니다.
* 장점: 계산이 빠르고, 대규모 데이터셋에 적합합니다.
* 단점: K(군집 수)를 미리 지정해야 하며, 비구형 군집이나 밀도 차이가 큰 군집을 잘 처리하지 못합니다.
* 사용 예: 고객 세분화, 이미지 압축 등.

2. Hierarchical Clustering
* 특징: 데이터 포인트 간의 거리 행렬을 사용하여 계층적으로 군집을 형성합니다. 데이터 포인트를 개별 군집으로 시작하여 점차적으로 병합하거나, 모든 데이터 포인트를 하나의 군집으로 시작하여 점차적으로 분할합니다.
* 장점: 군집 수를 사전에 지정할 필요가 없습니다. 덴드로그램을 통해 군집 형성 과정을 시각화할 수 있습니다.
* 단점: 계산 복잡도가 높아 대규모 데이터셋에는 부적합할 수 있습니다.
* 사용 예: 유전자 데이터 분석, 문서 클러스터링 등.

3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
* 특징: 밀도 기반 군집 분석 알고리즘으로, 밀도가 높은 영역에서 군집을 형성합니다. 노이즈 포인트를 감지하고 군집으로 포함시키지 않습니다.
* 장점: 군집의 모양에 제약이 없으며, 노이즈를 효과적으로 처리합니다.
* 단점: 적절한 하이퍼파라미터(eps, min_samples)를 선택하는 것이 어려울 수 있습니다.
* 사용 예: 지리적 데이터 분석, 이상 탐지 등.

4. Mean Shift
* 특징: 데이터 포인트의 밀도 중심으로 이동하며 군집을 형성합니다. 군집의 수를 사전에 지정할 필요가 없습니다.
* 장점: 다양한 군집 형태를 효과적으로 탐지할 수 있습니다.
* 단점: 계산 복잡도가 높아 대규모 데이터셋에는 부적합할 수 있습니다.
* 사용 예: 이미지 세분화, 컴퓨터 비전 등.

5. Gaussian Mixture Models (GMM)
* 특징: 데이터가 여러 개의 가우시안 분포로 구성된다고 가정합니다. 각 데이터 포인트가 특정 가우시안 분포에 속할 확률을 계산합니다.
* 장점: 군집의 모양이 가우시안 분포를 따르는 경우 효과적입니다.
* 단점: 계산이 복잡하며, 초기화에 민감할 수 있습니다.
* 사용 예: 이상 탐지, 데이터 분포 모델링 등.

6. Spectral Clustering
* 특징: 데이터의 유사성 행렬을 기반으로 하는 군집 분석 알고리즘입니다. 그래프 이론을 사용하여 데이터 포인트를 군집으로 나눕니다.
* 장점: 비선형 구조를 효과적으로 처리할 수 있습니다.
* 단점: 대규모 데이터셋에는 부적합할 수 있습니다.
* 사용 예: 커뮤니티 탐지, 이미지 분할 등.

7. OPTICS (Ordering Points To Identify the Clustering Structure)
* 특징: DBSCAN과 유사하지만 다양한 밀도에서도 군집을 효과적으로 찾을 수 있습니다.
* 장점: 밀도가 변하는 데이터셋에 적합합니다.
* 단점: 파라미터 설정이 어렵습니다.
* 사용 예: 지리적 데이터 분석, 이상 탐지 등.

8. BIRCH (Balanced Iterative Reducing and Clustering using Hierarchies)
* 특징: 대규모 데이터셋을 효과적으로 처리하기 위한 군집 분석 알고리즘입니다. 데이터 포인트를 점진적으로 요약하여 군집 구조를 형성합니다.
* 장점: 메모리 효율적이며 대규모 데이터셋에 적합합니다.
* 단점: 적절한 하이퍼파라미터 설정이 필요합니다.
* 사용 예: 대규모 데이터셋의 실시간 군집 분석.

9. Affinity Propagation
* 특징: 데이터 포인트 간의 "메시지 전달"을 통해 군집을 형성합니다. 각 데이터 포인트가 다른 포인트를 대표로 선택하여 군집 중심이 됩니다.
* 장점: 군집 수를 미리 지정할 필요가 없습니다.
* 단점: 계산 복잡도가 높으며 대규모 데이터셋에는 부적합할 수 있습니다.
* 사용 예: 이미지 처리, 문서 클러스터링 등.

10. HDBSCAN (Hierarchical Density-Based Spatial Clustering of Applications with Noise)
* 특징: DBSCAN의 확장 버전으로, 다양한 밀도 군집을 찾을 수 있습니다. 계층적 군집 분석과 밀도 기반 군집 분석의 장점을 결합합니다.
* 장점: 군집 수를 미리 지정할 필요가 없으며, 밀도가 변하는 데이터셋에 적합합니다.
* 단점: 설정 파라미터가 많고 복잡할 수 있습니다.
* 사용 예: 밀도가 다양한 데이터셋의 군집 분석.

11. Fuzzy C-Means Clustering
* 특징: 각 데이터 포인트가 하나의 군집에 속하는 것이 아니라 여러 군집에 소속될 수 있습니다. 소속도 확률을 기반으로 군집을 형성합니다.
* 장점: 데이터 포인트가 여러 군집에 걸쳐 있을 수 있는 경우에 유용합니다.
* 단점: 군집 수를 미리 지정해야 하며, 계산이 복잡할 수 있습니다.
* 사용 예: 이미지 처리, 패턴 인식 등.

12. Self-Organizing Maps (SOM)
* 특징: 인공신경망의 일종으로, 고차원 데이터를 저차원 격자 형태로 표현합니다. 비선형 데이터 구조를 효과적으로 학습할 수 있습니다.
* 장점: 데이터의 복잡한 구조를 시각적으로 표현할 수 있습니다.
* 단점: 설정 파라미터가 많고, 학습이 느릴 수 있습니다.
* 사용 예: 시각화, 차원 축소, 데이터 탐색 등.

```python
# Gap_Statistic - OptimalK 대체 오류 이유 모르겠음... 패키지 업데이트로 인하여 꼬인듯?


import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt

def gap_statistic(X, n_refs=20, max_clusters=10):
    gaps = np.zeros((max_clusters,))
    results_list = []

    for k in range(1, max_clusters + 1):
        ref_disps = np.zeros(n_refs)

        for i in range(n_refs):
            random_reference = np.random.random_sample(size=X.shape)
            km = KMeans(n_clusters=k, n_init=10, random_state=42)  # n_init 설정
            km.fit(random_reference)
            ref_disp = np.mean(np.min(cdist(random_reference, km.cluster_centers_, 'euclidean'), axis=1))
            ref_disps[i] = ref_disp

        km = KMeans(n_clusters=k, n_init=10, random_state=42)  # n_init 설정
        km.fit(X)
        orig_disp = np.mean(np.min(cdist(X, km.cluster_centers_, 'euclidean'), axis=1))

        gap = np.log(np.mean(ref_disps)) - np.log(orig_disp)
        gaps[k - 1] = gap

        results_list.append({'clusterCount': k, 'gap': gap})

    resultsdf = pd.DataFrame(results_list)
    return gaps, resultsdf

# X_df를 사용하여 Gap Statistic 계산
gaps, resultsdf = gap_statistic(X_df.to_numpy(), n_refs=20, max_clusters=10)

# Gap Statistic 시각화
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(gaps) + 1), gaps, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Gap Statistic')
plt.title('Gap Statistic for Different Number of Clusters')
plt.xticks(range(1, len(gaps) + 1))
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, MeanShift, SpectralClustering, AffinityPropagation
from sklearn.mixture import GaussianMixture
from sklearn.cluster import Birch, OPTICS
from hdbscan import HDBSCAN
from fcmeans import FCM
from minisom import MiniSom
import matplotlib.pyplot as plt

optimal_clusters = 4

# 모델 리스트
models = {
    'KMeans': KMeans(n_clusters=optimal_clusters, n_init=10, random_state=2024),
    'Hierarchical': AgglomerativeClustering(n_clusters=optimal_clusters),
    'DBSCAN': DBSCAN(eps=0.5, min_samples=5),  # DBSCAN은 군집 수를 지정하지 않음
    'MeanShift': MeanShift(),  # MeanShift는 군집 수를 지정하지 않음
    'GMM': GaussianMixture(n_components=optimal_clusters, random_state=2024),
    'SpectralClustering': SpectralClustering(n_clusters=optimal_clusters, random_state=2024),
    'OPTICS': OPTICS(min_samples=5),  # OPTICS는 군집 수를 지정하지 않음
    'BIRCH': Birch(n_clusters=optimal_clusters),
    'AffinityPropagation': AffinityPropagation(random_state=2024),  # AffinityPropagation은 군집 수를 지정하지 않음
    'HDBSCAN': HDBSCAN(min_cluster_size=5),  # HDBSCAN은 군집 수를 지정하지 않음
    'FCM': FCM(n_clusters=optimal_clusters),
    'SOM': MiniSom(optimal_clusters, 1, X_df.shape[1], sigma=0.3, learning_rate=0.5)
}

# 실루엣 스코어 저장용 딕셔너리
silhouette_scores = {}

# 희소 행렬을 밀집 배열로 변환
X_dense = X_df.to_numpy()

# 모델 학습 및 평가
for model_name, model in models.items():
    try:
        if model_name == 'GMM':
            model.fit(X_dense)
            labels = model.predict(X_dense)
        elif model_name == 'FCM':
            model.fit(X_dense)
            labels = model.predict(X_dense)
        elif model_name == 'SOM':
            model.train_random(X_dense, 100)
            labels = np.array([model.winner(x)[0] for x in X_dense])
        else:
            model.fit(X_dense)
            labels = model.labels_

        # DBSCAN과 OPTICS의 경우, 모든 데이터 포인트가 군집에 포함되지 않을 수 있으므로, 실루엣 점수 계산 시 주의 필요
        if len(set(labels)) > 1:
            score = silhouette_score(X_dense, labels)
        else:
            score = -1  # 군집이 1개 이하인 경우 실루엣 점수를 -1로 설정
    except Exception as e:
        score = -1
        print(f"Error in {model_name}: {e}")

    silhouette_scores[model_name] = score

# 실루엣 스코어 출력
for model_name, score in silhouette_scores.items():
    print(f'{model_name}: {score}')

# 실루엣 스코어 시각화
plt.figure(figsize=(10, 8))
plt.barh(list(silhouette_scores.keys()), list(silhouette_scores.values()), color='skyblue')
plt.xlabel('Silhouette Score')
plt.title('Silhouette Score by Clustering Algorithm')
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import SpectralClustering
from sklearn.metrics import make_scorer, silhouette_score
from sklearn.model_selection import GridSearchCV

# 'Cluster' 열 제거 (필요한 경우)
if 'Cluster' in X_df.columns:
    X_df = X_df.drop(columns=['Cluster'])

def spectral_silhouette_score(estimator, X):
    labels = estimator.fit_predict(X)
    if len(set(labels)) <= 1:  # 모든 데이터가 하나의 클러스터에 속하는 경우
        return -1
    return silhouette_score(X, labels)

# SpectralClustering 모델 및 하이퍼파라미터 그리드 설정 (n_clusters를 4로 고정)
spectral = SpectralClustering(n_clusters=4, random_state=42)
param_grid = {
    'affinity': ['nearest_neighbors', 'rbf'],
    'assign_labels': ['kmeans', 'discretize'],
    'n_neighbors': [5, 10, 20],  # affinity='nearest_neighbors'일 때만 사용
    'gamma': [0.01, 0.1, 1]  # affinity='rbf'일 때만 사용
}

# GridSearchCV 설정
grid_search = GridSearchCV(spectral, param_grid, scoring=make_scorer(spectral_silhouette_score), cv=3, n_jobs=-1)
grid_search.fit(X_df)

# 최적의 파라미터와 실루엣 점수 출력
print("Best parameters found: ", grid_search.best_params_)
print("Best silhouette score: ", grid_search.best_score_)

# 최적의 SpectralClustering 모델로 클러스터 예측 및 데이터프레임에 추가
best_spectral = grid_search.best_estimator_
X_df['Spectral_Labels'] = best_spectral.fit_predict(X_df)

print(X_df.head())
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import SpectralClustering
from sklearn.metrics import silhouette_score

# 'Spectral_Labels' 열 제거 (필요한 경우)
if 'Spectral_Labels' in X_df.columns:
    X_df = X_df.drop(columns=['Spectral_Labels'])

# 최적의 파라미터를 찾기 위한 수동 설정
n_clusters = 4
best_score = -1
best_params = None
best_labels = None

# 하이퍼파라미터 조합 시도
param_grid = {
    'affinity': ['nearest_neighbors', 'rbf'],
    'assign_labels': ['kmeans', 'discretize'],
    'n_neighbors': [5, 10, 20],  # affinity='nearest_neighbors'일 때만 사용
    'gamma': [0.01, 0.1, 1]  # affinity='rbf'일 때만 사용
}

for affinity in param_grid['affinity']:
    for assign_labels in param_grid['assign_labels']:
        if affinity == 'nearest_neighbors':
            for n_neighbors in param_grid['n_neighbors']:
                try:
                    spectral = SpectralClustering(
                        n_clusters=n_clusters,
                        affinity=affinity,
                        assign_labels=assign_labels,
                        n_neighbors=n_neighbors,
                        random_state=42
                    )
                    labels = spectral.fit_predict(X_df)
                    score = silhouette_score(X_df, labels)
                    if score > best_score:
                        best_score = score
                        best_params = {
                            'affinity': affinity,
                            'assign_labels': assign_labels,
                            'n_neighbors': n_neighbors
                        }
                        best_labels = labels
                except Exception as e:
                    print(f"Error with params affinity={affinity}, assign_labels={assign_labels}, n_neighbors={n_neighbors}: {e}")
        elif affinity == 'rbf':
            for gamma in param_grid['gamma']:
                try:
                    spectral = SpectralClustering(
                        n_clusters=n_clusters,
                        affinity=affinity,
                        assign_labels=assign_labels,
                        gamma=gamma,
                        random_state=42
                    )
                    labels = spectral.fit_predict(X_df)
                    score = silhouette_score(X_df, labels)
                    if score > best_score:
                        best_score = score
                        best_params = {
                            'affinity': affinity,
                            'assign_labels': assign_labels,
                            'gamma': gamma
                        }
                        best_labels = labels
                except Exception as e:
                    print(f"Error with params affinity={affinity}, assign_labels={assign_labels}, gamma={gamma}: {e}")

print("Best parameters found: ", best_params)
print("Best silhouette score: ", best_score)

# 최적의 SpectralClustering 모델로 클러스터 예측 및 데이터프레임에 추가
X_df['Spectral_Labels'] = best_labels

print(X_df['Spectral_Labels'].value_counts())
print(X_df.head())
```

```python
# Retention 분석 / 특정 기간 후에도 재방문한 사용자 비율

AARRR_df['event_date'] = AARRR_df['event_time'].dt.date
first_purchase = AARRR_df[AARRR_df['event_type'] == 'purchase'].groupby('user_id')['event_date'].min().reset_index(name='first_purchase_date')
AARRR_df = AARRR_df.merge(first_purchase, on='user_id', how='left')
AARRR_df['retention'] = AARRR_df.apply(lambda row: row['event_date'] >= row['first_purchase_date'] if pd.notnull(row['first_purchase_date']) else False, axis=1)

retention = AARRR_df[AARRR_df['retention']].groupby(['first_purchase_date', 'event_date'])['user_id'].nunique().reset_index()
retention_pivot = retention.pivot(index='first_purchase_date', columns='event_date', values='user_id').fillna(0)
retention_pivot.head()
```

```python
# Revenue 분석 / 사용자별 매출을 분석

revenue = AARRR_df[AARRR_df['event_type'] == 'purchase'].groupby('user_id')['price'].sum().reset_index(name='total_revenue')
revenue.head()
```

```python
# Referral 분석 / 추천 이벤트를 분석합니다. 데이터셋에 해당 이벤트가 포함되어 있는 경우만 수행
if 'referral' in AARRR_df['event_type'].unique():
    referral = AARRR_df[AARRR_df['event_type'] == 'referral'].groupby(AARRR_df['event_time'].dt.date)['user_id'].nunique().reset_index(name='referrals')
else:
    referral = pd.DataFrame(columns=['event_date', 'referrals'])
referral.head()
```

```python
# 결과 합치기 및 시각화 준비

final_results = pd.DataFrame()
final_results['date'] = pd.to_datetime(acquisition['event_time']).dt.date
final_results['new_users'] = acquisition['new_users']
final_results['activated_users'] = activation['activated_users']
final_results['total_revenue'] = revenue['total_revenue']

if not referral.empty:
    final_results['referrals'] = referral['referrals']

# Save to CSV for Tableau import
final_results.to_csv('/mnt/data/aarr_results.csv', index=False)

import ace_tools as tools; tools.display_dataframe_to_user(name="AARRR Analysis Results", dataframe=final_results)

final_results.head()
```
