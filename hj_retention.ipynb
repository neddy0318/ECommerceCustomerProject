```python
df['event_time'] = pd.to_datetime(df['event_time'])
retention_base = df[['user_id', 'new_user_session(15m)', 'event_time']].drop_duplicates()
retention_base['date_ym'] = retention_base['event_time'].dt.to_period('M')
retention_base
```

```python
date_ym_list = sorted(list(retention_base['date_ym'].unique()))
```

```python
from tqdm import tqdm

# 데이터프레임을 date_ym과 user_id로 groupby하여 사용자 집합을 미리 계산
user_sets = retention_base.groupby('date_ym')['user_id'].apply(set).to_dict()

retention = pd.DataFrame()

for s in tqdm(date_ym_list):
    period_start_users = user_sets.get(s, set())
    for t in date_ym_list:
        if s <= t:
            period_target_users = user_sets.get(t, set())
            retained_users = period_start_users.intersection(period_target_users)
            retention_rate = len(retained_users) / len(period_start_users) if period_start_users else 0
            temp = pd.DataFrame({'cohort': [s], 'date_ym': [t], 'retention_rate': [retention_rate]})
            retention = pd.concat([retention, temp])
```

```python
retention
```

```python
retention['cohort_size(month)'] = retention.apply(lambda x: (x['date_ym'] - x['cohort']).n, axis=1)
retention
```

```python
retention_matrix = retention.pivot(index='cohort', columns='date_ym', values='retention_rate')

plt.figure(figsize=(12, 8))
sns.heatmap(retention_matrix, annot=True, fmt=".1%", cmap="YlGnBu")
plt.title('Retention Matrix')
plt.show()
```

# 일주일 단위 리텐션
- 구매 주기가 긴 경우, 월간 리텐션보다 주간 리텐션이 더 세밀하게 사용자 행동 패턴을 파악할 수 있음

## 구매 완료 데이터

```python
retention_base_purchase['date_ymw'] = retention_base_purchase['event_time'].dt.to_period('W')
retention_base_purchase['date_ymw_str'] = retention_base_purchase['date_ymw'].astype(str)
retention_base_purchase
```

```python
user_sets_weekly = retention_base_purchase.groupby('date_ymw')['user_id'].apply(set).to_dict()

date_ymw_list = sorted(list(retention_base_purchase['date_ymw'].unique()))

retention_weekly = []


for i, s in enumerate(tqdm(date_ymw_list)):
    period_start_users = user_sets_weekly.get(s, set())
    if period_start_users:
        for t in date_ymw_list[i:]:
            period_target_users = user_sets_weekly.get(t, set())
            retained_users = period_start_users.intersection(period_target_users)
            retention_rate = len(retained_users) / len(period_start_users)
            retention_weekly.append({'cohort': s, 'date_ymw_str': t, 'retention_rate': retention_rate})
            

retention_weekly_df = pd.DataFrame(retention_weekly)
          
```

```python
retention_weekly_df
```

```python
import matplotlib.dates as mdates   

retention_pivot = retention_weekly_df.pivot(index='cohort', columns='date_ymw_str', values='retention_rate')

plt.figure(figsize=(12,8))
sns.heatmap(retention_pivot, annot=True, fmt=".2f", cmap="Blues")

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%W'))
plt.title("Weekly Retention Heatmap with Purchased data")
plt.show()
```

- 첫 주 이후의 주차별로 리텐션이 급격히 감소함
